### Train - test split of the dataset

```r
train<-sample(c(rep(0,0.7*nrow(inflation_final)),
                rep(1,0.3*nrow(inflation_final))))
train_inflation<-inflation_final[train==0,]
test_inflation<-inflation_final[train==1,]

```

# Modelling the data

### Multiple linear regression:

```r
lm_inflation <- lm(inflation ~ electricity+ gasPrice + petrolPrice + 
                   dieselPrice,data=train_inflation)
                   
summary(lm_inflation)
```

#### Finding the mean square error:
Means squar error tells how robust your model is.

```r
test_data<-data.frame(petrolPrice=(test_inflation$petrolPrice),
                      dieselPrice=(test_inflation$dieselPrice),
                     gasPrice=(test_inflation$gasPrice),
                     electricity=(test_inflation$electricity) 
                    )
pred_inflation<-predict(lm_inflation,test_data,interval="prediction")
actual_inflation<-test_inflation$inflation
```

#### Root mean square value for the our model is:

```r
rmse<-round(mean((actual_inflation-pred_inflation)^2,digits=2))
message("The root mean square error value for the model is:",rmse)
```

### Stepwise linear regression:

```r
steplm_inflation<-step(lm_inflation,direction="both",trace="true")
summary(steplm_inflation)
```
